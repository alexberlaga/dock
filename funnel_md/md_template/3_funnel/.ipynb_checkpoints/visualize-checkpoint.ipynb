```python
import nglview as nv
import mdtraj as md
import matplotlib.pyplot as plt
import numpy as np
def slice(traj, selection):
    return traj.atom_slice(traj.top.select(selection))
```

```python
m_traj = md.load('md_whole.xtc', top='md_whole.gro')
m_traj = slice(m_traj, 'not water and not name NA and not name CL')
m_traj.center_coordinates()
ref_lig = md.load('../a_insert/lig.pdb')
lig_nres = ref_lig.top.n_residues
total_nres = m_traj.top.n_residues
prot_nres = total_nres - lig_nres
t1 = m_traj.atom_slice(m_traj.top.select(f"resid 0 to {prot_nres - 1}"))
t2 = m_traj.atom_slice(m_traj.top.select(f"resid >= {prot_nres}"))
```

```python
m_view = nv.NGLWidget()
m_view.add_trajectory(t1)
m_view.add_trajectory(t2)
for v, rep in zip(m_view, ['ribbon', 'ball+stick']):
    v.clear_representations()
    v.add_representation(rep, color='residueIndex')
# m_view = nv.show_mdtraj(m_traj)
# create an nglview object and add the trajectory to it
# view = nv.show_mdtraj(m_traj)
# m_view.clear()
# m_view.add_ball_and_stick()
# display the view
m_view
```

```python
rmsf_per_ps = np.zeros(int(len(m_traj.xyz) / 10) + 1)
times = np.arange(int(len(m_traj.xyz) / 10) + 1) / 500
for i in range(0, len(m_traj.xyz), 10):
    pos = m_traj.xyz[:i]
    means = np.mean(pos, axis=0)
    sq_dev = np.square(pos - means)
    rmsf = np.sqrt(np.sum(sq_dev))
    times[int(i/10)] = i / 500
    rmsf_per_ps[int(i/10)] = rmsf * 500 / pos.shape[0] 
    
plt.plot(times, rmsf_per_ps)
plt.show()
```

```python
times = .002 * np.arange(m_traj.n_frames)
psis = np.zeros((4, m_traj.n_frames))
f = plt.figure(figsize=(25,5))
for i in range(1, 4):
    selection = "resid " + str(i) + " and name NL or resid " + str(i) + " and name CA or resid " + str(i) + " and name CLP or resid " + str(i+1) + " and name NL" 
    psis[i-1] = md.compute_dihedrals(m_traj, [m_traj.topology.select(selection)]).T
axs = f.subplots(1, 3)
for i in range(3):
    axs[i].plot(times, 180/np.pi * psis[i], "x")
    axs[i].set(xlabel="Time (ps)", ylabel="$\psi_" + str(i+1) + "$ Angle (degrees)")
plt.show()
```

```python
times = .002 * np.arange(m_traj.n_frames)
phis = np.zeros((4, m_traj.n_frames))
f = plt.figure(figsize=(25,5))
for i in range(1, 4):
    selection = "resid " + str(i-1) + " and name CLP or resid " + str(i) + " and name NL or resid " + str(i) + " and name CA or resid " + str(i) + " and name CLP"
    phis[i-1] = md.compute_dihedrals(m_traj, [m_traj.topology.select(selection)]).T
axs = f.subplots(1, 3)
for i in range(3):
    axs[i].plot(times, 180/np.pi * phis[i], "x")
    axs[i].set(xlabel="Time (ps)", ylabel="$\phi_" + str(i+1) + "$ Angle (degrees)")
plt.show()
```

```python
def rmsf(t):
    tt = t
    if t == 0 or tt > len(phis[0]):
        tt = len(phis[0])
    sin_phis = np.sin(phis[:, :tt])
    cos_phis = np.cos(phis[:, :tt])
    sin_psis = np.sin(psis[:, :tt])
    cos_psis = np.cos(psis[:, :tt])
    mean_sin_phi = np.mean(sin_phis, axis=1)
    mean_sin_psi = np.mean(sin_psis, axis=1)
    mean_cos_phi = np.mean(cos_phis, axis=1)
    mean_cos_psi = np.mean(cos_psis, axis=1)
    sin_phi_dev = np.square(sin_phis - mean_sin_phi[:, np.newaxis])
    sin_psi_dev = np.square(sin_psis - mean_sin_psi[:, np.newaxis])
    cos_phi_dev = np.square(cos_phis - mean_cos_phi[:, np.newaxis])
    cos_psi_dev = np.square(cos_psis - mean_cos_psi[:, np.newaxis])
    sin_phi_rmsf = np.sqrt(np.mean(sin_phi_dev, axis=1))
    sin_psi_rmsf = np.sqrt(np.mean(sin_psi_dev, axis=1))
    cos_phi_rmsf = np.sqrt(np.mean(cos_phi_dev, axis=1))
    cos_psi_rmsf = np.sqrt(np.mean(cos_psi_dev, axis=1))
    return np.array([sin_phi_rmsf, sin_psi_rmsf, cos_phi_rmsf, cos_psi_rmsf])

f = plt.figure(figsize=(35,15))
axs = f.subplots(4, 3)

rmsfs = np.zeros((int(len(phis[0]) / 100) + 1, 4, 3))
times = np.arange(0, len(phis[0]), 100) / 500

for i in range(0, len(phis[0]), 100):
    rmsfs[int(i/100), :, :] = rmsf(i)[:, :3]

for i in range(4):
    for j in range(3):
        axs[i, j].plot(times, rmsfs[:, i, j])
            
plt.show()    
```
